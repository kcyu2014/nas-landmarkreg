```python
# process sparse rank.
def sparse_rank_by_simple_bin(ranks, perfs, threshold=1e-2):
    new_ranks = []
    c_perf = 0.0
    c_rank = -1
    if threshold < 1.:
        multipler = 1 / threshold
    
    for r, p in zip(ranks, perfs):
        p_rank = p * multipler // 1
        cp_rank = c_perf * multipler // 1
        if p_rank > cp_rank:
            c_rank += 1
        new_ranks.append(c_rank)
        c_perf = p
        print(f'map {r} to {c_rank}')
```

```python
# rectify the old model_ids by the new one.
def rectify_model_ids_byvalid_to_bytest(model_ids, nasbench, hash_perfs=hash_perfs):
    hashs = nasbench.hash_rank
    perfs = [hash_perfs[h] for h in hashs]
    hashs_bytest, perfs_bytest = sort_hash_perfs(hashs, [p[1] for p in perfs])
    hashs_byvalid, _ = sort_hash_perfs(hashs, [p[0] for p in perfs])
    rect_model_ids = [hashs_bytest.index(hashs_byvalid[i]) for i in model_ids]
    return rect_model_ids, [perfs_bytest[i] for i in rect_model_ids]

# function begins
def compute_sparse_kendalltau(model_ids, model_perfs, gt_perfs, threshold=1e-4, 
                              fn_map_perf_to_new_rank=nasbench_utils.sparse_rank_by_simple_bin):
    avg_perfs_multiplier = 1e-2 if 1 < np.average(model_perfs) < 100 else 1.
    model_perfs = [p * avg_perfs_multiplier for p in model_perfs]

    gt_perfs,_ = sort_hash_perfs(gt_perfs, model_perfs)
    # print(model_perfs[:10])
    # obtain the new sparse ground-truth rank
    sgt_model_ids, sgt_perfs = sort_hash_perfs(model_ids, gt_perfs, verbose=False)
    sgt_sparse_ranks = fn_map_perf_to_new_rank(sgt_model_ids, sgt_perfs, threshold=threshold, verbose=True)

    #current version, just support this simple version, which is, project the sparse gt model ids.
    pred_sparse_ranks = [sgt_sparse_ranks[sgt_model_ids.index(i)] for i in model_ids]


    # pred_sparse_ranks = fn_map_perf_to_new_rank(model_ids, model_perfs, threshold=threshold, verbose=True)
    print("Reduced ranks from {} to {}".format(len(set(model_ids)), len(set(gt_sparse_ranks))))
    print(kendalltau(sgt_sparse_ranks, pred_sparse_ranks))

```
